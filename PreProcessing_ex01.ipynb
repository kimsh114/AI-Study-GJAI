## 데이터 전처리는 중요한 파트이다.

평균은 0이 되고 표준편차가 1이되는 것이 표준 정규분포이다.

```python
import seaborn as sns

iris = sns.load_dataset('iris')
iris
```

```python
x = iris.iloc[:,:-1]
x
```

```python
from sklearn.preprocessing import scale
x_scaled = scale(x)
x_scaled[:5,:]  #표준화 작업을 하면 numpy  배열로 반환된다.
```

```python
x_scaled.mean(axis=0) # 표준화 한 데이터들의 평균
```

```python
for scaled_mean in x_scaled.mean(axis=0):
 print("{:10.9f}".format(scaled_mean)) #사실상 0에 가까운 수
```

```python
from sklearn.preprocessing import minmax_scale
iris_minmax_scaled = minmax_scale(x)
iris_minmax_scaled[:5,:]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
iris_scaled = sc.fit_transform(iris.iloc[:,:-1])
iris_scaled[:5,:]
```

```python
iris_origin = sc.inverse_transform(iris_scaled)
iris_origin[:5,:]
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
iris_scaled =  mms.fit_transform(iris.iloc[:,:-1])
iris_scaled[:-5,:]
```

### 레이블 인코딩
- 실제 값에 상관없이 0~k -1까지의 정수로 변환하는 것이다.
- fit 과 transform은 반드시 짝을 지어서 사용한다. 

```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
y = iris.species # 문자형태를 숫자로 바꿔야하니깐 바꿔야 하는 것이 바람직한데, 필수는 아니다.
y
```

```python
le = LabelEncoder()
le.fit(y) #fit은 학습의 개념이다. y 를 계산함(학습함)
species = le.transform(y)
species
```

```python
le.classes_ #어떤 것을 변환 시켰는지 알려주는 것
```

### 타이타닉 데이터 중 성별과 emabarked를 원핫인코딩으로 변환하자

# 결측값 처리

```python
import seaborn as sns
iris = sns.load_dataset('iris')
x = iris.iloc[:,:-1]
x
```

```python
import random
for col in range(4):
    x.iloc[[random.sample(range(len(iris)),10)],col] = float('nan')
x.head()
```

```python
x.mean(axis = 0)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

imp_mean = SimpleImputer(missing_values = np.nan, strategy = 'median')
imp_mean.fit_transform(x)[0:5,:]
```

## titanic 에서 결측값을 채우기

```python
import pandas as pd
df = pd.read_csv('./titanic/train.csv')
df
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

imp_mean = SimpleImputer(missing_values = np.nan, strategy = 'mean')
df_age = imp_mean.fit_transform(df[['Age']])
df_age
```

```python
imp_mean = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent') #most_frequent => 최빈값
df_embarked = imp_mean.fit_transform(df[['Embarked']])
df_embarked
```

```python
df['Age'] = df_age
df['Embarked'] = df_embarked
df
```
