# - 케라스
- 딥러닝 모델은 은닉층이 두 단계 이상이 있는 것이 딥러닝이다.
- 손실값을 통해서 가중치와 바이어스를 최적의 값으로 업데이트 한다.
- 케라스의 기본 아키텍쳐는 모델을 기본으로 하고 있다.
- 딥러닝 모델은 신경망을 이용한 모델이라고 생각하면 된다

```python
# 라이브러리 임포트 및 학습 데이터 정의
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD
import numpy as np

x_data = np.array([1, 2, 3, 4, 5, 6])
t_data = np.array([3, 4, 5, 6, 7, 8]) 
print(tf.__version__) 
```

```python
# 모델 (model) 구축
model = Sequential() # 모델
model.add(Flatten(input_shape=(1,))) # 입력층(1차원을 입력함)
model.add(Dense(1, activation='linear')) # 출력층
# model.add(Dense(1, input_shape=(1,), activation='linear')) # 입력층 + 출력층
```

```python
# 모델 (model) 컴파일 및 summary
# 확률적 경사하강법인 SGD 를 옵티마이저로 설정
model.compile(optimizer=SGD(learning_rate=1e-2), loss='mse')
model.summary() 
```

```python
# 모델 학습
hist = model.fit(x_data, t_data, epochs=1000) # epochs=> 모든 데이터에 대해서 한 번 수행 했을 때를 1 epochs라고 한다.
```

```python
# 모델 평가 및 예측
result = model.predict(np.array([-3.1, 3.0, 3.5, 15.0, 20.1]))
print(result)
```

```python
# 차트 그리기
import matplotlib.pyplot as plt

plt.title('Loss Trend')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid()

plt.plot(hist.history['loss'], label = 'train loss')
plt.legend(loc = 'best')

plt.show()
```

```python
model.save('test.pkl')
```
